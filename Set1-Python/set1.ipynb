```python
def englishCheck(byteResult, threshold=0.99):
    printableCount=0
    total=len(byteResult)

    for b in byteResult:
        c = chr(b)
        if c in string.ascii_letters + string.digits + string.punctuation + string.whitespace:
            printableCount += 1
        else:
            # Check for non-printable ASCII
            if b < 32 or b > 126:
                return False

    return printableCount >= total * threshold
```

```python
def score(bstr: bytes):
  # https://en.wikipedia.org/wiki/Letter_frequency
  freq = {
      'a': 0.08167,
      'b': 0.01492,
      'c': 0.02782,
      'd': 0.04253,
      'e': 0.12702,
      'f': 0.02228,
      'g': 0.02015,
      'h': 0.06094,
      'i': 0.06966,
      'j': 0.00153,
      'k': 0.00772,
      'l': 0.04025,
      'm': 0.02406,
      'n': 0.06749,
      'o': 0.07507,
      'p': 0.01929,
      'q': 0.00095,
      'r': 0.05987,
      's': 0.06327,
      't': 0.09056,
      'u': 0.02758,
      'v': 0.00978,
      'w': 0.02360,
      'x': 0.00150,
      'y': 0.01974,
      'z': 0.00074,
      ' ': 0.13000,
  }
  return sum([freq.get(chr(b), 0) for b in bstr.lower()])
```

```python
def HammingDistance(str1:bytes, str2:bytes):
    count=0
    
    for b1,b2 in zip(str1,str2):
        count+=bin(b1^b2).count("1")
    
    return count
```

```python
def getBestKeySize(binaryData, minkeySize=2, maxKeySize=40):
    keydist=[]

    for keySize in range(minkeySize, maxKeySize):
        scores=[]

        for i in range(0, len(binaryData)//keySize-1):
            b1=binaryData[i*keySize : (i+1)*keySize]
            b2=binaryData[(i+1)*keySize:(i+2)*keySize]

            score=HammingDistance(b1,b2)/keySize

            scores.append(score)

        keydist.append((keySize, sum(scores) / len(scores)))
    
    keydist.sort(key=lambda x: x[1])
    return keydist[:3]
```

```python
def xorKey(bstr: bytes, key: int):
  return bytes([b ^ key for b in bstr])
```

```python
def bestScore(binaryData, top=3):
        results = [(i, score(xorKey(binaryData, i)), xorKey(binaryData, i)) for i in range(0, 256)]
        results.sort(reverse=True, key=lambda x: x[1])
        return results[:top]
```

```python
def crackRepeatingKeyXor(binaryData, minkeySize=2, maxKeySize=40, top_n=3):

    keySizes=getBestKeySize(binaryData, minkeySize, maxKeySize)
    print(keySizes)

    result=[]

# trying different key sizes
    for keySize, _ in keySizes:
        key=[]
        print(f"trying key size {keySize}")
# keySize number of blocks
        for i in range(keySize):

            block=binaryData[i::keySize]

            k, sc, res= bestScore(block, 1)[0]

            key.append(k)

        plaintext=repeatingKeyXor(binaryData, key)

        keyPlaintext=bytes(key).decode("utf-8", errors="ignore")

        result.append((score(plaintext),plaintext, key))

    return sorted(result, reverse=True, key=lambda x:x[0])[:top_n]
        

```

```python
#inverse column mix layer, had to chatgpt this
def gmul(a, b):
    p = 0
    for _ in range(8):
        if b & 1:
            p ^= a
        hi_bit_set = a & 0x80
        a = (a << 1) & 0xFF
        if hi_bit_set:
            a ^= 0x1B
        b >>= 1
    return p

def inverseMixColumnLayer(matrix):
    for c in range(4):  # for each column
        a = [matrix[r][c] for r in range(4)]
        matrix[0][c] = gmul(a[0], 0x0E) ^ gmul(a[1], 0x0B) ^ gmul(a[2], 0x0D) ^ gmul(a[3], 0x09)
        matrix[1][c] = gmul(a[0], 0x09) ^ gmul(a[1], 0x0E) ^ gmul(a[2], 0x0B) ^ gmul(a[3], 0x0D)
        matrix[2][c] = gmul(a[0], 0x0D) ^ gmul(a[1], 0x09) ^ gmul(a[2], 0x0E) ^ gmul(a[3], 0x0B)
        matrix[3][c] = gmul(a[0], 0x0B) ^ gmul(a[1], 0x0D) ^ gmul(a[2], 0x09) ^ gmul(a[3], 0x0E)
    return matrix
```

```python
#inverse row shift

def inverseShiftRowLayer(matrix):
    
    matrix[1]=matrix[1][-1:]+matrix[1][:-1]
    
    matrix[2]=matrix[2][-2:]+matrix[2][:-2]

    matrix[3]=matrix[3][-3:]+matrix[3][:-3]

    return matrix;



```

```python
def matrixMaker(byteLine):
    result=[[0]*4 for _ in range(4)]

    for i in range(0,16):
        result[i%4][i//4]=byteLine[i]
    
    return result
```

```python
def keyAddition(matrix, key):
    for row in range(4):
        for col in range(4):
            matrix[row][col]=matrix[row][col]^key[row][col]

    return matrix
```
